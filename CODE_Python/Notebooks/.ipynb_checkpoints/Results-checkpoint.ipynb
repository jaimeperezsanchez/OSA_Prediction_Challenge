## Regression Results

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Assign spreadsheet filename to `file`
file = 'Regression_Results.xlsx'
# Load spreadsheet
xl = pd.ExcelFile(file)
# Print the sheet names
print(xl.sheet_names)
```

```python
df_raw = xl.parse('Raw')
df_poly = xl.parse('Poly')
df_Standard = xl.parse('Standard')
df_MinMax = xl.parse('MinMax')
```

### RAW data 

```python
df_raw.set_index("Model (RAW data)", inplace=True)
df_raw
```

```python
df_raw = df_raw.sort_values('RMSE')
plt.figure(figsize=(12,9))
ax = sns.barplot(df_raw.index, df_raw.RMSE, alpha=0.8)
ax.set_xticklabels(df_raw.index, rotation='vertical', fontsize=10)
plt.show()
```

### Polynomial

```python
df_poly.set_index("Model (Poly)", inplace=True)
df_poly
```

```python
df_poly = df_poly.sort_values('RMSE')
plt.figure(figsize=(12,9))
ax = sns.barplot(df_poly.index, df_poly.RMSE, alpha=0.8)
ax.set_xticklabels(df_poly.index, rotation='vertical', fontsize=10)
plt.show()
```

### Standard 

```python
df_Standard.set_index("Model (Standard)", inplace=True)
df_Standard
```

```python
df_Standard = df_Standard.sort_values('RMSE')
plt.figure(figsize=(12,9))
ax = sns.barplot(df_Standard.index, df_Standard.RMSE, alpha=0.8)
ax.set_xticklabels(df_Standard.index, rotation='vertical', fontsize=10)
plt.show()
```

```python
df_Standard = df_Standard.sort_values('R²', ascending=False)
plt.figure(figsize=(12,9))
ax = sns.barplot(df_Standard.index, df_Standard['R²'], alpha=0.8)
ax.set_xticklabels(df_Standard.index, rotation='vertical', fontsize=10)
plt.show()
```

### MinMax

```python
df_MinMax.set_index("Model (MinMax)", inplace=True)
df_MinMax
```

```python
df_MinMax = df_MinMax.sort_values('RMSE')
plt.figure(figsize=(12,9))
ax = sns.barplot(df_MinMax.index, df_MinMax.RMSE, alpha=0.8)
ax.set_xticklabels(df_MinMax.index, rotation='vertical', fontsize=10)
plt.show()
```

## Classification Results

```python
# Assign spreadsheet filename to `file`
file = 'Classification_Results.xlsx'
# Load spreadsheet
xl = pd.ExcelFile(file)
# Print the sheet names
print(xl.sheet_names)
```

```python
df_raw = xl.parse('Raw')
df_poly = xl.parse('Poly')
df_Standard = xl.parse('Standard')
df_MinMax = xl.parse('MinMax')
```

### RAW data 

```python
df_raw.set_index("Model (RAW data)", inplace=True)
df_raw
```

```python
df_raw = df_raw.sort_values('F1_score',ascending=False)
plt.figure(figsize=(12,9))
ax = sns.barplot(df_raw.index, df_raw.F1_score, alpha=0.8)
ax.set_xticklabels(df_raw.index, rotation='vertical', fontsize=10)
plt.show()
```

```python
df_raw = df_raw.sort_values('Recall',ascending=False)
plt.figure(figsize=(12,9))
ax = sns.barplot(df_raw.index, df_raw.Recall, alpha=0.8)
ax.set_xticklabels(df_raw.index, rotation='vertical', fontsize=10)
plt.show()
```

### Polynomial

```python
df_poly.set_index("Model (Poly)", inplace=True)
df_poly
```

```python
df_poly = df_poly.sort_values('F1_score',ascending=False)
plt.figure(figsize=(12,9))
ax = sns.barplot(df_poly.index, df_poly.F1_score, alpha=0.8)
ax.set_xticklabels(df_poly.index, rotation='vertical', fontsize=10)
plt.show()
```

### Standard 

```python
df_Standard.set_index("Model (Standard)", inplace=True)
df_Standard
```

```python
df_Standard = df_Standard.sort_values('F1_score',ascending=False)
plt.figure(figsize=(12,9))
ax = sns.barplot(df_Standard.index, df_Standard.F1_score, alpha=0.8)
ax.set_xticklabels(df_Standard.index, rotation='vertical', fontsize=10)
plt.show()
```

### MinMax

```python
df_MinMax.set_index("Model (MinMax)", inplace=True)
df_MinMax
```

```python
df_MinMax = df_MinMax.sort_values('F1_score',ascending=False)
plt.figure(figsize=(12,9))
ax = sns.barplot(df_MinMax.index, df_MinMax.F1_score, alpha=0.8)
ax.set_xticklabels(df_MinMax.index, rotation='vertical', fontsize=10)
plt.show()
```
